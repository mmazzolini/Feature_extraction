### PCA FEATURE SELECTION

```python
from it_matrix import create_it_matrix
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import gaussian_kde

from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.compose import TransformedTargetRegressor
from sklearn.model_selection import GridSearchCV,TimeSeriesSplit
from sklearn.metrics import mean_squared_error

import seaborn as sns
import os

import pdb
```

```python
#choose file from daily_input/old : on these dataset spatial averaging wasn't executed.
path=r'C:\Users\mmazzolini\OneDrive - Scientific Network South Tyrol\Documents\conda\daily_input\old\\'
# ## Feature selection
daily_input = pd.read_csv(path+'CANDOGLIA_TOCE_2000_2019.csv', index_col=0, parse_dates=True)

#daily_clim = daily_climatology(daily_input)
it_matrix = create_it_matrix(daily_input, 12,30)

#drop the target variable
X = it_matrix.drop(columns=['Q'])

#SCALE THE VARIABLES
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#EXECUTE THE PCA ANALYSIS (100 components are considered enough)
pca = PCA(n_components=100)
results = pca.fit_transform(X_scaled)
```

```python
# ORGANIZE RESULTS IN A DATAFRAME WITH PROPER DIMENSIONS SO THAT WE CAN EXTRACT INFORMATION EASILY. 

var_t=[]
pixel=[]
past_months=[]

y= []
x= []

cols= np.array(X.columns)
for c in cols:
    #print(c)
    var_t.append(c[0])
    p=c.split('x')[:][1].split('y')
    pixel.append(p)
    
    x.append(int(p[0])) 
    y.append(int(p[1].split('_')[0]))
    
    nm=c.split('_')[1][1:]
    if nm=='':
        past_months.append('0')
    else : past_months.append(nm)
        

arrays = [var_t,past_months,x,y]
tuples = list(zip(*arrays))
index = pd.MultiIndex.from_tuples(tuples, names=["var", "past_periods","x","y"])

components=pd.DataFrame(pca.components_.transpose(),index=index)
```

Play with the principal_component variable to see how much explained variance every pixel contributes to each principal components

## TEMPERATURE

```python
principal_component=0
variable='T'


q=f'var == "{variable}"'
all_data=components.query(q)[principal_component]
hue_lim=(all_data.min(),all_data.max())

fig, ax = plt.subplots(figsize=(25,19))         # Sample figsize in inches
for i in range(12):
    q=f'var == "{variable}" and past_periods=="{i}"'
    data=components.query(q)[principal_component]

    data=data.reset_index().rename(columns={principal_component:'comp'})

    plt.subplot(3,4,i+1)
    sns.scatterplot(x='x',y='y', data=data,marker=",", hue='comp',s=700,hue_norm=hue_lim)
    plt.title(f'P.C.{principal_component}, var={variable}, {i} past periods')
    #plt.legend(False)
```

### CONCLUSION:
with regards to the last plot:
it seems clear that a low number of principal components can well represent the variance of the dataset.
From following experiments it will come up that using just the spatial statistics of a certain variable in a certain month(or period) can lead to a proper prediction, in a very similar way as PCA applied before the SVR would do.


In general we see that the SNOW variable is the one that exhibits the larger difference in explained variance in the spatial dimension. (the left-upper pixels represent the high elevation areas of the CANDOGLIA basin)
Most likely lower pixels (center+bottom right) experience a few snow events, hence they don'y contribute much to the explained variance of the first principal components. (play with the principal component variable for checking this.

The fact that for most of the variables the contribution to the first principal component is veery similar for all the pixels makes us think that the spatial variablity is not that important.
Hence we'll decide that the final models will be trained with spatial statistics. (we'll see results are very similar to PCA preprocessed input matrix and to giving as input all the pixels.)
